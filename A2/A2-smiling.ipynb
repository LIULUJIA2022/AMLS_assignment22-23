## a) Feature Extraction Function
使用dlib库的get_frontal_face_detector模块探测出人脸，然后使用shape_predictor_68_face_landmarks.dat特征数据预测人脸特征数值。
Mouth: [49,68] 用于判断人脸是否有微笑

```python
# 68-points dlib预测器

#dlib特征提取器
detector = dlib.get_frontal_face_detector()  #使用dlib库的get_frontal_face_detector模块探测出人脸
#功能：人脸检测画框
#参数：无
#返回值：默认的人脸检测器

#加载并初始化检测器
#dlib的68点特征预测器，进行人脸 68点特征提取：
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat') #预测器
#利用Dlib官方训练好的模型“shape_predictor_68_face_landmarks.dat”进行68点标定

# how to find frontal human faces in an image using 68 landmarks.  These are points on the face such as the corners of the mouth, along the eyebrows, on the eyes, and so forth.
# The face detector we use is made using the classic Histogram of Oriented
# Gradients (HOG) feature combined with a linear classifier, an image pyramid,
# and sliding window detection scheme.  The pose estimator was created by
# using dlib's implementation of the paper:
# One Millisecond Face Alignment with an Ensemble of Regression Trees by
# Vahid Kazemi and Josephine Sullivan, CVPR 2014
# and was trained on the iBUG 300-W face landmark dataset (see https://ibug.doc.ic.ac.uk/resources/facial-point-annotations/):
#     C. Sagonas, E. Antonakos, G, Tzimiropoulos, S. Zafeiriou, M. Pantic.
#     300 faces In-the-wild challenge: Database and results.
#     Image and Vision Computing (IMAVIS), Special Issue on Facial Landmark Localisation "In-The-Wild". 2016.
```

## b)Bounding box to match the faces
take a bounding predicted by dlib and convert it to the format (left, top, right, bottom) 

```python
def rect_to_bounding_box(rect):
    left = rect.left()
    top = rect.top()
    right = rect.right() - left
    bottom = rect.bottom() - top
    
    # print("Detection {}: Left: {} Top: {} Right: {} Bottom: {}".format(i, left, top, right, bottom))
    # print(left, top, right, bottom)

    # return a tuple of (left, top, right, bottom)
    return (left, top, right, bottom)
```

## c)Extract landmarks at face region
load the image, detect the landmarks of the face, and then return the image and the landmarks

```python
def run_dlib_shape(image):
    dlibout = []
    dlibout_mouth = []
    # load the input image, resize it to uint8 type, and convert it to grayscale
    resized_image = image.astype('uint8')

    gray = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
    
    # 再次将灰度图片转化成uint8 type
    gray = gray.astype('uint8')
    
    # detect faces in the grayscale image
    rects = detector(gray, 1)
    
    # 返回的结果是包含着检测到的人脸的方框的左上角坐标和右下角坐标
    #print(rects)
    
    num_faces = len(rects)

    # 未检测到人脸
    if num_faces == 0:
        dlibout = None
        # print('Cannot dectect a face, wrong image!')
    
    else:
        # 检测到人脸
        face_areas = np.zeros((1, num_faces))
        face_shapes = np.zeros((136, num_faces), dtype = np.int64)

        # loop over the face detections
        for (i, rect) in enumerate(rects):
            # determine the facial landmarks for the face region, then
            # convert the facial landmark (x, y)-coordinates to a NumPy array
            # 得到关键点的坐标
            temp_shape = predictor(gray, rect)
            # 此处返回的shape并不是关键点的坐标，我们需要将其中包含的坐标信息提取出来。
            temp_shape = shape_to_np(temp_shape, dtype=  'int')
            # 将shape的返回值转换为能用的坐标信息
        
            # convert dlib's rectangle to a OpenCV-style bounding box
            #  (x, y, w, h) = face_utils.rect_to_bounding_box(rect)
            (left, top, right, bottom) = rect_to_bounding_box(rect)
            face_shapes[:, i] = np.reshape(temp_shape, [136])
            face_areas[0, i] = right * bottom
        # find largest face and keep
        dlibout = np.reshape(np.transpose(face_shapes[:, np.argmax(face_areas)]), [68, 2])
        
        # location of 68 facial regions: mouth is [49:68]
        dlibout_mouth = dlibout[49:68,:]

    return dlibout_mouth, resized_image
```

## d) Convert facial landmarks into Numpy array
将shape的返回值转换为能输入进模型的坐标信息

```python
def shape_to_np(shape, dtype="int"):
    # initialize the list of (x, y)-coordinates
    # 创建68*2
    coords = np.zeros((shape.num_parts, 2), dtype=dtype)

    # loop over all facial landmarks and convert them into a 2-tuple of (x, y)-coordinates
    for i in range(0, shape.num_parts):
        coords[i] = (shape.part(i).x, shape.part(i).y) # 第i个关键点的横纵坐标
    # print(coords)
    # return the list of (x, y)-coordinates
    return coords
```

## e) Extract the facial landmarks and smiling label all images in celeba dataset
This funtion extracts the landmarks features for all images in the folder 'dataset_AMLS_22-23/celeba/img'
 
It also extracts the smiling label for each image for all labels in the file 'dataset_AMLS_22-23/celeba/'labels.csv''
    
    :return:
        landmark_features:  an array containing 68 landmark points for each image in which a face was detected
        smiling_labels:     an array containing the smiling label (non-smiling=0 and smiling=1) for each image in
                            which a face was detected
        right_img:          detecting 68 landmark points in the images correctly                    
        wrong_img:          error detecting 68 landmark points in the images due to non-frontal faces           

```python
def extract_features_labels(img_dir, celeba_dir, labels_filename):
   
    # load the img_dir to list all images file name
    image_paths = [os.path.join(img_dir, l) for l in os.listdir(img_dir)]
    target_size = None
    # load the labels file
    labels_file = open(os.path.join(celeba_dir, labels_filename), 'r')
    # list all images smiling label
    lines = labels_file.readlines() 
    # gender_labels = {line.split('\t')[0] : int(line.split('\t')[2]) for line in lines[1:]}
    smiling_labels = {line.split('\t')[0] : int(line.split('\t')[3]) for line in lines[1:]}
    if os.path.isdir(img_dir):
        # 能识别到人脸的图片
        right_img = []
        
        face_features = []
        smilings = []
        # smile = []
        
        # 未能识别到人脸的图片
        wrong_img = []
        
        for img_path in image_paths:
            file_name= img_path.split('.')[0].split('/')[-1]

            # load image
            img = image.img_to_array(
                image.load_img(img_path,
                               target_size=target_size,
                               interpolation='bicubic'))
            features, _ = run_dlib_shape(img)
            
            if features is not None:
                right_img.append(file_name)
                face_features.append(features)
                smilings.append(smiling_labels[file_name])
                # print('the detector can recognize the face', file_name)
            else:
                wrong_img.append(file_name)
                # print('the detector can recognize the face', file_name)

    landmark_features = np.array(face_features)
    simling_labels = (np.array(smilings) + 1)/2 # simply converts the -1 into 0, so not smile=0 and smile=1
    # return 带红点的正确图名字，特征点坐标，性别标签，错误图名字
    return landmark_features, smiling_labels, right_img, wrong_img
```

Using these features is optional, since there is a variety of facial landmarks detectors you could choose from. Also the same dlib framework can be used to train other facial landmark detectors or custom shape predictors.

Note that we will use features extracted from images rather than images themselves.

```python
def get_data(X, Y):
    Y = np.array([Y, -(Y - 1)]).T
    # Shuffle and split the data into training(70%) and test(30%) set 
    X, Y = shuffle(X, Y)
    tr_X, te_X, tr_Y, te_Y = train_test_split(X, Y, train_size = 0.7, random_state = 42)
    
    # Reshape training and test X into (n_samples, n_features)
    tr_X = tr_X.reshape(tr_X.shape[0], tr_X.shape[1]*tr_X.shape[2])
    te_X = te_X.reshape(te_X.shape[0], te_X.shape[1]*te_X.shape[2])
    
    # Unzipped training and test Y from (n_samples,) into (n_samples)
    tr_Y = list(zip(*tr_Y))[0]
    te_Y = list(zip(*te_Y))[0]
    
    return tr_X, te_X, tr_Y, te_Y
print(Y)
# input the landmarks, gender(output of extract_features_labels() to get_data
# feature as X,label as Y
tr_feature, te_feature, tr_label, te_label = get_data(landmarks, smiling)
#  tr_label and te_label are tuples
print(tr_feature.shape)
print(te_feature.shape)
```

## a)(ii) Fit SVM model to the training data
Fit 3 different Support Vector Classifier (SVC) such as polynomial, rbf and linear kernel to the training data

class **sklearn.svm.SVC(*, C=1.0, kernel='rbf', degree=3, gamma='scale', coef0=0.0, shrinking=True, probability=False, tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=-1, decision_function_shape='ovr', break_ties=False, random_state=None)**

## a)(iii) Fit SVM model with polynomial, rbf and linear kernel to the training data
### Experiment with different kernels

use the following 3 different kernel functions implemented in the scikit learn library:
1. linear: \begin{align} <x,x'> \end{align}

2. polynomial: \begin{align} (\gamma <x,x'> +r)^d\end{align} 

3. rbf:\begin{align}e^{(-\gamma||x-x'||^2)} \end{align} 

```python
# Linear kernel SVM
def SVM_linear(x_train, y_train, x_test):
    model = SVC(C = 1.0, kernel = 'linear', degree = 3, gamma = 'scale')
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    return y_pred

# Polynomial kernel SVM
def SVM_poly(x_train, y_train, x_test, d):
    model = SVC(C = 1.0, kernel = 'poly', degree = 3, gamma = 'scale')
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    return y_pred

# Radial basis function kernel SVM
def SVM_rbf(x_train, y_train, x_test):
    model = SVC(C = 1.0, kernel = 'rbf', degree = 3, gamma = 'scale')
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    return y_pred 

# Result of classifier for linear SVM
y_pred = SVM_linear(tr_feature, tr_label, te_feature)
print('Linear kernel SVC')
print()
print('Accuracy of classifier: ', accuracy_score(te_label,y_pred))
print()
print('Confusion matrix: \n', confusion_matrix(te_label,y_pred))
print()
print(classification_report(te_label,y_pred))

# Result of classifier for polynomial SVM
y_pred2 = SVM_poly(tr_feature, tr_label, te_feature, 3)
print('Polynomial kernel SVC')
print()
print('Accuracy of classifier: ', accuracy_score(te_label,y_pred2))
print()
print('Confusion matrix: \n', confusion_matrix(te_label,y_pred2))
print()
print(classification_report(te_label,y_pred2))
              
# Result of classifier for RBF SVM
y_pred3 = SVM_rbf(tr_feature, tr_label, te_feature)
print('RBF kernel SVC')
print()
print('Accuracy of classifier: ', accuracy_score(te_label,y_pred3))
print()
print('Confusion matrix: \n', confusion_matrix(te_label,y_pred3))
print()
print(classification_report(te_label,y_pred3))
```

# 3) Hyperparameter tuning for SVC (Support Vector Classifiers)

## a) SVC with linear kernel
\begin{align} <x,x'> \end{align}

Hyperparameter is only **C**

```python
# Set the parameters by cross-validation
tuned_parameters = [{'kernel': ['linear'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}]

scores = ['precision', 'recall']

for score in scores:
    print("Tuning hyper-parameters for SVC with linear kernel %s" % score)
    print()

    clf = GridSearchCV(SVC(), tuned_parameters, cv=5,
                       scoring='%s_macro' % score)
    clf.fit(tr_feature, tr_label)

    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("Grid scores on development set:")
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"
              % (mean, std * 2, params))
    print()

    print("Detailed classification report:")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = te_label, clf.predict(te_feature)
    print(classification_report(y_true, y_pred))
    print()
```
